# Data preparing for Naive Bayes Classification 

### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(['Male'],axis=1)
y=df['Male']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.30, random_state=101)
```

# Training Model

### Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

gnb=GaussianNB()

gnb.fit(X_train,y_train)
```

```python
prediction = gnb.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,prediction))
```

```python

print('Train of accuracy of NBC alog',gnb.score(X_train, y_train))
```

### Multinomial Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
mnb=MultinomialNB()
mnb.fit(X_train,y_train)
```

```python
prediction=mnb.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,prediction))
```

### Bernoulli Naive Bayes

```python
from sklearn.naive_bayes import BernoulliNB
bnb=BernoulliNB()
bnb.fit(X_train,y_train)
```

```python
predcition=bnb.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predcition))
```

### Complement Naive Bayes

```python
from sklearn.naive_bayes import ComplementNB
cnb=ComplementNB()
cnb.fit(X_train,y_train)

prediction=cnb.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,prediction))
```
